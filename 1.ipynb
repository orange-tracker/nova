```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 데이터 로딩
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42
)

# 정확도 저장용 리스트
train_scores = []
test_scores = []

# k값 1 ~ 10까지 반복
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))

# 그래프 그리기
plt.plot(range(1, 11), train_scores, label='Train accuracy')
plt.plot(range(1, 11), test_scores, label='Test accuracy')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.legend()
plt.title('k 1-10 accuracy')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# 데이터 불러오기
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42
)

# 정확도 저장 리스트
train_scores = []
test_scores = []

# k 값을 1부터 10까지 바꿔가며 반복
for k in range(1, 11):
    kn = KNeighborsClassifier(n_neighbors=k)
    kn.fit(X_train, y_train)
    train_scores.append(kn.score(X_train, y_train))
    test_scores.append(kn.score(X_test, y_test))

# 결과 출력
for k in range(1, 11):
    print(f"k={k}: Train Accuracy={train_scores[k-1]:.3f}, Test Accuracy={test_scores[k-1]:.3f}")

# 그래프 그리기
plt.plot(range(1, 11), train_scores, label='Train Accuracy')
plt.plot(range(1, 11), test_scores, label='Test Accuracy')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.title('k-NN Classifier Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
